```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_neural_network(layers):
    """
    Draws a neural network diagram using the specified layer structure.

    Parameters:
    layers (list of int): A list where each element represents the number of neurons in that layer.

    Returns:
    None
    """
    # Create a directed graph
    G = nx.DiGraph()

    # Add nodes for each layer
    for i in range(len(layers)):
        for j in range(layers[i]):
            G.add_node((i, j), label=f"$y_{{{j}}}^{{{i}}}$")  # Add label with layer index

            if i > 0:
                for k in range(layers[i - 1]):
                    G.add_edge((i - 1, k), (i, j))  # Add edge with weight

    # Position the nodes in a grid layout with vertical alignment
    pos = {}
    max_layer_width = max(layers)
    for i in range(len(layers)):
        layer_height = layers[i]
        vertical_offset = (max_layer_width - layer_height) / 2
        for j in range(layers[i]):
            pos[(i, j)] = (i, max_layer_width - j - 1 - vertical_offset)

    # Draw the nodes with labels inside
    node_labels = nx.get_node_attributes(G, 'label')
    nx.draw(G, pos, labels=node_labels, with_labels=True, node_size=1000, 
            node_color="white", edgecolors="black", edge_color="gray", font_size=10)

    # Draw edge labels (weights w_ji)
    edge_labels = {}
    for edge in G.edges(data=True):
        start, end, weight = edge
        j, i = start[1], end[1]  # j is the input neuron, i is the output neuron
        start_layer = start[0]  # Layer index of the input neuron
        end_layer = end[0]      # Layer index of the output neuron
        edge_labels[(start, end)] = f"$w_{{{j}{i}}}^{{{end_layer}{start_layer}}}$"  # Label format as w_ji^out, in

    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=12)

    plt.title("Neural Network Architecture")
    plt.show()
```

```python
# Define the architecture of the neural network 
layers = [3, 1]  # Example

# Draw the neural network
draw_neural_network(layers)
```

```python
# Define the architecture of the neural network 
layers = [4, 5, 3, 1]  # Example

# Draw the neural network
draw_neural_network(layers)
```
