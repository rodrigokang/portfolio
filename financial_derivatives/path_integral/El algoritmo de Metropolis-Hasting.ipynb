# El algoritmo de Metropolis-Hasting

```python
# Importamos las librer√≠as
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funci√≥n de densidad de probabilidad que queremos muestrear (distribuci√≥n normal en este caso)
def TargetDistribution(x):
    return np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

# Funci√≥n de propuesta (distribuci√≥n normal en este caso)
def ProposalDistribution(x, sigma=0.5):
    return np.random.normal(x, sigma)

# Algoritmo de Metropolis-Hastings
def MetropolisHastings(iterations, initial_value):
    samples = [initial_value]

    for _ in range(iterations):
        current_sample = samples[-1]
        
        # Propuesta de nuevo valor
        proposed_sample = ProposalDistribution(current_sample)
        
        # Raz√≥n de aceptaci√≥n
        acceptance_ratio = TargetDistribution(proposed_sample) / TargetDistribution(current_sample)
        
        # Aceptar la propuesta con probabilidad igual a la raz√≥n de aceptaci√≥n
        if np.random.uniform(0, 1) < acceptance_ratio:
            samples.append(proposed_sample)
        else:
            samples.append(current_sample)

    return np.array(samples)

# Par√°metros
iterations = 10000
initial_value = 0.0

# Obtener muestras usando Metropolis-Hastings
samples = MetropolisHastings(iterations, initial_value)

# Visualizar resultados
plt.hist(samples, bins=50, density=True, alpha=0.5, color='blue', label='Muestras')
x = np.linspace(-5, 5, 100)
plt.plot(x, TargetDistribution(x), color='red', label='Distribuci√≥n objetivo')
plt.title('Metropolis-Hastings')
plt.legend()
plt.show()
```
