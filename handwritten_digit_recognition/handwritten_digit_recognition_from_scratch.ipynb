```python
import gzip
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(filename):
    """
    Load data from a gzip-compressed pickle file.
    
    Args:
        filename (str): The path to the gzip-compressed pickle file.
        
    Returns:
        data: The unpickled data.
    """
    with gzip.open(filename, 'rb') as f:
        data = pickle.load(f, encoding='latin1')
    return data

def load_mnist():
    """
    Load the MNIST dataset.
    
    Args:
        None
        
    Returns:
        Tuple: Three tuples containing training, validation, and test datasets respectively.
    """
    mnist_file = 'mnist.pkl.gz'  # Change 'mnist.pkl.gz' to the path of your downloaded file
    
    # Load the data
    mnist = load_data(mnist_file)
    
    # Unpack the data
    training_data, validation_data, test_data = mnist
    
    # Separate features and labels
    train_features, train_labels = training_data
    validation_features, validation_labels = validation_data
    test_features, test_labels = test_data
    
    # Reshape features into image matrices
    train_features = np.reshape(train_features, (-1, 28, 28))
    validation_features = np.reshape(validation_features, (-1, 28, 28))
    test_features = np.reshape(test_features, (-1, 28, 28))
    
    return (train_features, train_labels), (validation_features, validation_labels), (test_features, test_labels)

# Load the MNIST dataset
(train_features, train_labels), (validation_features, validation_labels), (test_features, test_labels) = load_mnist()
```

```python
# Print basic information about the loaded datasets
print("Training dataset shape (features):", train_features.shape)
print("Training dataset shape (labels):", train_labels.shape)
print("Validation dataset shape (features):", validation_features.shape)
print("Validation dataset shape (labels):", validation_labels.shape)
print("Test dataset shape (features):", test_features.shape)
print("Test dataset shape (labels):", test_labels.shape)
```

```python
# Plot a few examples of handwritten digits
num_examples = 5

fig, axes = plt.subplots(num_examples, num_examples, figsize=(8, 8))
for i in range(num_examples):
    for j in range(num_examples):
        index = i * num_examples + j
        axes[i, j].imshow(train_features[index], cmap='gray')
        axes[i, j].set_title(f"Label: {train_labels[index]}")
        axes[i, j].axis('off')

plt.tight_layout()
plt.show()
```
