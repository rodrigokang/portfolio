```python
# Importar librería para leer archivos
import os

# Agregar la ruta del del módulo modelos
import sys
sys.path.append('C:/Users/RJKANG/Desktop/modelos_cencosud/')

# Importar librerías
import pandas as pd
from datetime import datetime
import time

# Importar la función desde el script
from main_smk import main_rfm, get_parameters, check_records

# Nombre del analista y fecha de ejecución
analista = "rjkang"
ts_actualizacion = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
# Usuario y contraseña
user = 'rjkang'
password = '#Rjkang2023'
# Nombre del script input
script_name = 'in_rfm_smk.sql'

# Obtener todos los parámetros desde la base de datos
parametros_df = get_parameters(user, password)

# Carpeta local donde guardar los archivos
local_folder = 'C:/Users/RJKANG/Desktop/modelos_cencosud/smk/'

# Asegurarse de que la carpeta existe
os.makedirs(local_folder, exist_ok=True)

# Iterar sobre cada requerimiento y ejecutar el proceso
for idx, row in parametros_df.iterrows():
    requerimiento = row['requerimiento']
    date_start = row['date_start']
    date_end = row['date_end']
    cadena = row['cadena']
    marca = row['marca']
    region = row['region']
    canal = row['canal']
    nrolocal = row['local']
    tipo_cliente = row['tipo_cliente']
    area_solicitante = row['area_solicitante']
    solicitante = row['solicitante']
    
    # <nombre de la tabla>
    table_name = 'rfm_kmeans_smk_customer'
    
    # Verificar si hay registros para el requerimiento
    if check_records(table_name, requerimiento, user, password):
        continue
    
    # Iniciar el contador de tiempo en segundos
    start_time_seconds = time.time()
    
    # Ejecutar la función y obtener el DataFrame
    output = main_rfm(date_start, date_end, cadena, marca, region, canal, 
                      nrolocal, tipo_cliente, analista, ts_actualizacion, 
                      requerimiento, area_solicitante, solicitante,
                      user, password, script_name)
    
    # Finalizar el contador de tiempo en segundos y calcular la duración
    end_time_seconds = time.time()
    execution_time_seconds = end_time_seconds - start_time_seconds
    
    # Agregar columna de tiempo de ejecución al DataFrame
    output['exe_time_sec'] = execution_time_seconds
    
    # <><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><> #
    # Guardar datos localmente
    # <><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><><> #

    # Nombre del archivo
    file_name = f'rfm_kmeans_smk_customer_{requerimiento}.csv'
    # Ruta completa del archivo
    file_path = os.path.join(local_folder, file_name)

    # Guardar el DataFrame en la carpeta local
    output.to_csv(file_path, index=False)
    
    # Imprimir el tiempo de ejecución en segundos
    print("Tiempo de ejecución en segundos:", execution_time_seconds)

    # Calcular el tiempo de ejecución en minutos
    execution_time_minutes = execution_time_seconds / 60

    # Imprimir el tiempo de ejecución en minutos
    print("Tiempo de ejecución en minutos:", execution_time_minutes)
    
    # Imprimir mensaje de progreso
    print(f"Proceso finalizado para el requerimiento {requerimiento}. Archivo guardado en {file_path}.")
```
